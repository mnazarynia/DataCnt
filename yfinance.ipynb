```python
tickersymbole = ["LHA.DE", "NVDA", "TSLA"]

for symbol in tickersymbole:
    print(symbol)

    # 1. Hole Aktienkursdaten
    import yfinance
    ti = yfinance.Ticker( symbol )
    df = ti.history(start="2004-05-15", end="2024-05-15")

    # 2. Visualisiere den Tagesschlusskurs der Aktie als Linienplot
    import matplotlib.pyplot as plt
    import seaborn as sns
    sns.set(style="whitegrid")
    plt.figure(figsize=(17,2))
    plt.plot( df["Close"], color="gray" )
    plt.xlabel("Datum", color="red", fontsize=15)
    plt.ylabel("Tagesschlusskurs []", color="brown")
    plt.title("Aktienkurs der Aktie: " + symbol, color=(0.0, 0.0, 0.0) )
    sns.lineplot(x=df.index, y=df["Close"], color="gray")
    plt.show()
```

```python
import yfinance
import matplotlib.pyplot as plt
import seaborn as sns



tickersymbole = ["LHA.DE", "NVDA", "TSLA"]

for symbol in tickersymbole:
    print(symbol)


    ti = yfinance.Ticker(symbol)
    df = ti.history(start="2004-05-15", end="2024-05-15")
    sns.set(style="whitegrid")

    plt.figure(figsize=(17, 6))
    sns.lineplot(x=df.index, y=df["Close"], color="gray")
    plt.xlabel("Datum", color="red", fontsize=15)
    plt.ylabel("Tagesschlusskurs []", color="brown")
    plt.title(f"Aktienkurs der Aktie: {symbol}", color="black")
    plt.show()
```
